```python
import pandas as pd

df = pd.read_csv("Nassau Candy Distributor.csv")   # use your file name

# 1️⃣ Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

# 2️⃣ Create Lead Time
df['Lead Time'] = (df['Ship Date'] - df['Order Date']).dt.days

# 3️⃣ Check invalid values
print("Negative lead time:", (df['Lead Time'] < 0).sum())
print("More than 20 days:", (df['Lead Time'] > 20).sum())

# 4️⃣ Clean dataset
df = df[(df['Lead Time'] >= 0) & (df['Lead Time'] <= 20)]

print("Remaining rows:", len(df))
```

```python
# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

# FIX YEAR ISSUE (make same year)
df['Ship Date Fixed'] = df['Ship Date'].apply(lambda x: x.replace(year=2024))

# If ship date becomes earlier than order date, push to next year
df.loc[df['Ship Date Fixed'] < df['Order Date'], 'Ship Date Fixed'] = \
    df['Ship Date Fixed'] + pd.DateOffset(years=1)

# Calculate lead time
df['Lead Time'] = (df['Ship Date Fixed'] - df['Order Date']).dt.days

print(df['Lead Time'].describe())
```

```python
factory_map = {
"Wonka Bar - Nutty Crunch Surprise":"Lot's O' Nuts",
"Wonka Bar - Fudge Mallows":"Lot's O' Nuts",
"Wonka Bar -Scrumdiddlyumptious":"Lot's O' Nuts",

"Wonka Bar - Milk Chocolate":"Wicked Choccy's",
"Wonka Bar - Triple Dazzle Caramel":"Wicked Choccy's",

"Laffy Taffy":"Sugar Shack",
"SweeTARTS":"Sugar Shack",
"Nerds":"Sugar Shack",
"Fun Dip":"Sugar Shack",
"Fizzy Lifting Drinks":"Sugar Shack",

"Everlasting Gobstopper":"Secret Factory",
"Lickable Wallpaper":"Secret Factory",
"Wonka Gum":"Secret Factory",

"Hair Toffee":"The Other Factory",
"Kazookles":"The Other Factory"
}

df['Factory'] = df['Product Name'].map(factory_map)

print(df['Factory'].isna().sum())
df[['Product Name','Factory']].head(10)
```

```python
df['Route'] = df['Factory'] + " → " + df['State/Province']

df[['Factory','State/Province','Route']].head(10)
```

```python
route_performance = df.groupby('Route').agg(
    Shipments=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Variability=('Lead Time','std')
).reset_index()

route_performance = route_performance.sort_values(by='Avg_Lead_Time')

route_performance.head(10)
```

```python
factory_map = {
"Wonka Bar - Nutty Crunch Surprise":"Lot's O' Nuts",
"Wonka Bar - Fudge Mallows":"Lot's O' Nuts",
"Wonka Bar -Scrumdiddlyumptious":"Lot's O' Nuts",
"Wonka Bar - Milk Chocolate":"Wicked Choccy's",
"Wonka Bar - Triple Dazzle Caramel":"Wicked Choccy's",
"Laffy Taffy":"Sugar Shack",
"SweeTARTS":"Sugar Shack",
"Nerds":"Sugar Shack",
"Fun Dip":"Sugar Shack",
"Fizzy Lifting Drinks":"Sugar Shack",
"Everlasting Gobstopper":"Secret Factory",
"Lickable Wallpaper":"Secret Factory",
"Wonka Gum":"Secret Factory",
"Hair Toffee":"The Other Factory",
"Kazookles":"The Other Factory"
}

df['Factory'] = df['Product Name'].map(factory_map)
```

```python
df['Route'] = df['Factory'] + " → " + df['State/Province']
```

```python
threshold = df['Lead Time'].mean()
df['Delayed'] = df['Lead Time'] > threshold

route_kpi = df.groupby('Route').agg(
    Shipments=('Order ID','count'),
    Avg_Lead_Time=('Lead Time','mean'),
    Delay_Rate=('Delayed','mean')
).reset_index()

route_kpi = route_kpi.sort_values('Avg_Lead_Time')

route_kpi.head(10)
```

```python
route_kpi_filtered = route_kpi[route_kpi['Shipments'] >= 10]

route_kpi_filtered.head(10)
```

```python
# Best performing routes (high volume)
best_routes = route_kpi_filtered.sort_values('Avg_Lead_Time').head(5)
best_routes
```

```python
# Worst performing routes (high volume)
worst_routes = route_kpi_filtered.sort_values('Avg_Lead_Time', ascending=False).head(5)
worst_routes
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.barh(
    worst_routes['Route'], 
    worst_routes['Delay_Rate'],
    color="orange"   # Professional red
)

plt.xlabel("Delay Rate")
plt.title("Top 5 Worst Routes by Delay Rate", fontsize=14, fontweight='bold')

plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(11,6))

best_percent = best_routes['Delay_Rate'] * 100

plt.barh(
    best_routes['Route'],
    best_percent,
    color="#27ae60"   # Professional green
)

for i, v in enumerate(best_percent):
    plt.text(v + 1, i, f"{v:.1f}%", va='center', fontsize=10)

plt.xlabel("Delay Rate (%)")
plt.title("Top 5 Best Routes by Delay Rate", fontsize=15, fontweight='bold')

plt.gca().invert_yaxis()
plt.xlim(0,100)
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
